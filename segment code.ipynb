```python
def get_custom_stopwords(stop_words_file):
    with open(stop_words_file, "r",encoding="utf-8") as f:
        stopwords = f.read()
    stopwords_list = stopwords.split('\n')
    custom_stopwords_list = [i for i in stopwords_list]
    return custom_stopwords_list
```
